1. Definition (Say This in an Interview)

Linear Search (also called Sequential Search) is a searching algorithm that checks each element of a data structure one by one until the target element is found or the structure ends.

It does not require sorted data.

the main objective is to sort ands return a value 
if we find the element then return the index or else return -1

# 2. When to Use Linear Search

Use Linear Search when:

The data is unsorted

The dataset size is small

Only one search is required

Memory usage must be constant

Data is coming as a stream

Interview line:

“Since the array is unsorted and constraints are small, linear search is the most optimal approach.”

#Time and Space Complexity (Very Important)
Case	Time Complexity
1. Best Case	O(1)
2. Average Case	O(n)
3. Worst Case	O(n)

Space Complexity: O(1)

Why?
No extra memory is used.#

# the first main code example

```python
# the most basic code oi this is
def linear_search(arr,target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

```python
arr = [4,7,1,9,3]
target = 9
linear_search(arr,target)
```

# T1 Find First Occurrence (MOST COMMON)

Given an unsorted array, return the index of the first occurrence of the target.

. Interview Insight

. Use early exit

Do NOT scan full array once found

```python
def first_occurrence(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i # this is what we mean by early exit as the code will stop here
    return -1
```

time complexity = O(n)

# T2 Find Last Occurrence
Problem Statement

Given an array arr and a value target, return the index of the last occurrence of target.
If the target does not exist, return -1.

Interview Insight

No early exit

Keep updating index

```python
def linear_search(arr,target):
    last = -1
    for i in range(len(arr)):
        if arr[i] == target:
            last = i # this will keep the whole code going and upadate the value of last accordingly
    return last
```

Common Interview Follow-Ups
Q: What if the array is empty?

Answer:
Loop doesn’t run → returns -1.

Q: What if all elements are the target?

Answer:
The index keeps updating → last index returned.

Q: Can this be optimized further?

Answer:
No, because every element must be checked to ensure it is the last occurrence.

One-Line Interview Explanation (Memorize)

“I traverse the array completely and update the index each time the target is found. After the loop, the stored index represents the last occurrence.”

# T3 Count Frequency of an Element — Detailed Explanation
Problem Statement

Given an array arr and a value target, count how many times the target appears in the array.

If the target does not appear, return 0.

```python
def linear_search(arr,target):
    count = 0
    for i in range(len(arr)):
        if arr[i] == target:
            count +=1
    return count
```

# 5. Find All Indices of an Element
Problem Statement

Return a list of all indices where the target appears.

Interview Insight

Store results

No early exit

```python
def linear_search(arr, target):
    indice = []
    for i in range(len(arr)):
        if arr[i] == target:
            indice.append(i)
    return indice
```

# T5 Example: Count Character Frequency in a String



```python
def count_char(s, ch):
    count = 0
    for c in s:
        if c == ch:
            count += 1
    return count
```
