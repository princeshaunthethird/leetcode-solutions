1. Definition (Say This in an Interview)

Linear Search (also called Sequential Search) is a searching algorithm that checks each element of a data structure one by one until the target element is found or the structure ends.

It does not require sorted data.

the main objective is to sort ands return a value 
if we find the element then return the index or else return -1

##2. When to Use Linear Search

Use Linear Search when:

The data is unsorted

The dataset size is small

Only one search is required

Memory usage must be constant

Data is coming as a stream

Interview line:

“Since the array is unsorted and constraints are small, linear search is the most optimal approach.”

#Time and Space Complexity (Very Important)
Case	Time Complexity
Best Case	O(1)
Average Case	O(n)
Worst Case	O(n)

Space Complexity: O(1)

Why?
No extra memory is used.#

```python
# the most basic code oi this is
def linear_search(arr,target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

```python
arr = [4,7,1,9,3]
target = 9
linear_search(arr,target)
```
